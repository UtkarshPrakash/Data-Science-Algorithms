### Kaggle tutorial for BigQuery basics (SQL Course)

```python
from google.cloud import bigquery
```

```python
client = bigquery.Client() # Creating an object client

# Constructing reference to hacker_news dataset
dataset_ref = client.dataset("hacker_news", project="bigquery-public-data") 

# API request for fetching the dataset
dataset = client.get_dataset(dataset_ref)
```

```python
tables = list(client.list_tables(dataset)) # lists all tables

for table in tables:
    print(table.table_id)
```

```python
# Constructing reference to 'full' table
table_ref = dataset_ref.table("full")

# API request for fetching the table
table = client.get_table(table_ref)
```

```python
table.schema
```

```python
client.list_rows(table, max_results=6).to_dataframe()
```

```python
client.list_rows(table, selected_fields=table.schema[2:3], max_results=4).to_dataframe()
```
